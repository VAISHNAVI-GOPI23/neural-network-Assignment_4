```python
import torch, torch.nn as nn, torch.optim as optim
from torchvision import datasets, transforms
from torchvision.utils import make_grid
import matplotlib.pyplot as plt

# Hyperparams
latent_dim, batch_size, lr, epochs = 100, 256, 0.0002, 20

# Load MNIST
dataloader = torch.utils.data.DataLoader(
    datasets.MNIST('./data', train=True, download=True,
                   transform=transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])),
    batch_size=batch_size, shuffle=True)

# Generator
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 128), nn.ReLU(True),
            nn.Linear(128, 784), nn.Tanh()
        )

    def forward(self, z):
        return self.model(z).view(-1, 1, 28, 28)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(), nn.Linear(784, 128), nn.LeakyReLU(0.2),
            nn.Linear(128, 1), nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

# Init
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
G, D = Generator().to(device), Discriminator().to(device)
opt_G, opt_D = optim.Adam(G.parameters(), lr=lr), optim.Adam(D.parameters(), lr=lr)
loss_fn = nn.BCELoss()
fixed_noise = torch.randn(16, latent_dim, device=device)
g_losses, d_losses, samples = [], [], {}

# Train
for epoch in range(epochs + 1):
    for real, _ in dataloader:
        real = real.to(device)
        z = torch.randn(real.size(0), latent_dim, device=device)
        fake = G(z)

        # Train D
        D.zero_grad()
        loss_real = loss_fn(D(real), torch.ones(real.size(0), 1, device=device))
        loss_fake = loss_fn(D(fake.detach()), torch.zeros(real.size(0), 1, device=device))
        loss_D = loss_real + loss_fake
        loss_D.backward()
        opt_D.step()

        # Train G
        G.zero_grad()
        loss_G = loss_fn(D(fake), torch.ones(real.size(0), 1, device=device))
        loss_G.backward()
        opt_G.step()

    g_losses.append(loss_G.item())
    d_losses.append(loss_D.item())

    if epoch in [0, 10, 20]:
        with torch.no_grad():
            G.eval()
            samples[epoch] = G(fixed_noise).cpu()
            G.train()

# Plot losses
plt.figure(figsize=(10, 4))
plt.plot(g_losses, label='Generator Loss')
plt.plot(d_losses, label='Discriminator Loss')
plt.title("Losses")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# Show images
for e in [0, 10, 20]:
    grid = make_grid(samples[e], nrow=4, normalize=True)
    plt.figure(figsize=(4, 4))
    plt.title(f"Epoch {e}")
    plt.imshow(grid.permute(1, 2, 0).squeeze())
    plt.axis("off")
    plt.show()
```
